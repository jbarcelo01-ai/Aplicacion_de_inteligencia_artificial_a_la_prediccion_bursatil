```python
import pandas as pd

tickers_df = pd.read_csv("../../data/external/tickers.csv")

tickers = tickers_df["ticker"].tolist()

tickers
```

```python
import yfinance as yf
from pathlib import Path

output_dir = Path("../../data/raw/stocks")
output_dir.mkdir(parents=True, exist_ok=True)

for ticker in tickers:

    print(f"Descargando {ticker}...")

    df = yf.download(
        ticker,
        start="2000-01-01",
        auto_adjust=False,
        progress=False
    )

    if df.empty:
        print(f"No se encontraron datos para {ticker}")
        continue

    # Limpiar columnas
    df = df.reset_index()

    df.columns = [
        col[0] if isinstance(col, tuple) else col
        for col in df.columns
    ]

    df = df.rename(columns={"index": "date"})

    df.columns = (
        pd.Index(df.columns)
        .str.lower()
        .str.replace(" ", "_")
    )

    # Agregar ticker
    df["ticker"] = ticker

    # Guardar parquet
    output_path = output_dir / f"{ticker.lower()}.parquet"

    df.to_parquet(output_path, index=False)

    print(f"Guardado: {output_path}")
```

```python
import pandas as pd
from pathlib import Path

data_dir = Path("../../data/raw/stocks")

files = list(data_dir.glob("*.parquet"))

files
```

```python
dfs = []

for file in files:

    df = pd.read_parquet(file)

    # Si no guardaste ticker antes
    if "ticker" not in df.columns:
        df["ticker"] = file.stem.upper()

    dfs.append(df)

master_df = pd.concat(dfs, ignore_index=True)

master_df.head(-10)
```

```python
master_df.dtypes
```

```python
master_df = master_df[
    ["ticker"] + [col for col in master_df.columns if col != "ticker"]
]
```

```python
master_df = master_df.sort_values(["ticker", "date"])
master_df.head(-10)
```
